```python
'''
random
'''

arrival_times = []

population = [n for n in range(4,48)]
weights = []
for n in range(len(population)):
    weights.append(1)
    
n_hgv = 22

for n in range(n_hgv):
    
    value = choices(population, weights)

    arrival_times.append(value[0])
    
    
rand_cumulative = np.zeros(48)

for n in range(len(arrival_times)):
    rand_cumulative[arrival_times[n]] += 1
    
fig, ax = plt.subplots(figsize=[15,5])

x = [n for n in range(48)]

ax.bar(x, rand_cumulative)
```

```python
'''
perfect
'''

arrival_times = [n for n in range(4,48,2)]

perf_cumulative = np.zeros(48)

for n in range(len(arrival_times)):
    perf_cumulative[arrival_times[n]] += 1
    
fig, ax = plt.subplots(figsize=[15,5])

x = [n for n in range(48)]

ax.bar(x, perf_cumulative)
```
