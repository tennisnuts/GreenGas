```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from random import choices
from collections import Counter

from scipy.interpolate import make_interp_spline, BSpline
```

```python
# 100 200 and 300 kg/day
NREL = [28.81, 19.26, 15.64]
x_nrel = [100,200,300]
```

```python


ME = [
1545.9947759098814
,821.8227335430934
,592.156733093183
,471.21432277564185
,396.56113701375597
,347.80690324589636
,313.804179180239
,288.2916107088094
,268.6542865234665
,253.2025623873435]
ME = np.array(ME)/32.09
x_me = [n for n in range(1,11)]
x_me = np.array(x_me)*32.09
```

```python
fig, ax = plt.subplots(figsize=[15,5])

ax.plot(x_nrel, NREL)
ax.plot(x_me, ME)
ax.set_ylim(0)
```

```python
xnew = np.linspace(x_me.min(), x_me.max(), 300) 
spl = make_interp_spline(x_me, ME, k=5)
power_smooth = spl(xnew)

new_x = np.linspace(np.min(x_nrel), np.max(x_nrel), 300) 
new_spline = make_interp_spline(x_nrel, NREL, k=2)
nrel_y = new_spline(new_x)
```

```python
fig, ax = plt.subplots(figsize=[15,5])

ax.plot(xnew, power_smooth, label='Optimised Station', color='teal')
ax.plot(new_x, nrel_y, label='NREL Estimate', color='orangered')
ax.set_ylim(0)
ax.set_xlim(0)
ax.grid(lw=0.2)
ax.set_xlabel('Delivered kg per Day', fontsize=14)
ax.set_ylabel('Hydrogen Price \n per kg', fontsize=14, rotation=0, labelpad=60)
ax.legend()
```
