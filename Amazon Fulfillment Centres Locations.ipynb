```python
crs = {"init": "epsg:27700"}
df = pd.DataFrame(coords)
df.columns = ["Lat", "Lon"]
df
```

```python
geo_df = gpd.GeoDataFrame(df, crs="epsg:27700", geometry=geometry)
geo_df
```

```python
fig, ax = plt.subplots(figsize=[15, 15])

uk_map.plot(ax=ax, alpha=1, color="grey")
geo_df.plot(ax=ax, markersize=20, color="blue", marker="o")
```
